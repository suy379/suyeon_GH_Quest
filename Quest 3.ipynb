```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
CCTV_Seoul = pd.read_csv('C:\\CCTV_in_Seoul.csv', encoding = 'utf-8') #인코딩이 안되면 engine = 'python'으로 해보세요! C드라이브에 파일 넣는 게 제일 잘됨
CCTV_Seoul.head()
```

```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0] : '구별'}, inplace= True)
CCTV_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('C:\\population_in_Seoul.xls', 
                          header = 2, usecols = 'B, D, G, J, N', encoding = 'utf-8')
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={pop_Seoul.columns[0] : '구별',
                         pop_Seoul.columns[1] : '인구수',
                         pop_Seoul.columns[2] : '한국인',
                         pop_Seoul.columns[3] : '외국인',
                         pop_Seoul.columns[4] : '고령자'}, inplace= True)
pop_Seoul.head()
```

```python
pop_Seoul.drop([0],inplace=True)
pop_Seoul.head()
```

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2016년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2014년']) / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul.head(5)
```

```python
data_result=pd.merge(CCTV_Seoul,pop_Seoul,on="구별")
data_result.head(5)
```

```python
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result.head()
```

# Quest 1 정답

```python
import numpy as np
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
f1 = np.poly1d(fp1)

data_result['외국인비율']=(data_result['외국인']/data_result['인구수'])*100
data_result['고령자비율']=(data_result['고령자']/data_result['인구수'])*100
data_result['오차']=np.abs(data_result['소계']-f1(data_result['인구수']))
data_result.set_index('구별', inplace=True)
data_result.head()
```

```python
result=data_result.sort_values(by = '오차', ascending = False)
```

```python
result.head()
```

# Quest 2 정답

```python
plt.figure(figsize=(15,10))
x=result['인구수']
y=result['소계']
a=result['오차']
plt.scatter(x,y, s=50,c=a)
fx = np.linspace(100000, 700000, 1000)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color ='r')
plt.xlabel('인구수')
plt.ylabel('소계')

for i in range(10):
    plt.text(x[i]*1.02,y[i]*0.98,result.index[i],fontsize=15)

plt.grid()
plt.colorbar()
plt.show()
#다음의 그래프는 x축은 인구수, y축은 소계를 나타낸 그래프이다.
#오차 칼럼 기준을 추가하여 데이터를 시각화해 보니, 아까 구했던 회귀직선과의 거리가 가장 먼 점들에 해당하는 index(구이름)를 알 수 있었다.(편의상 10개만)
#회귀직선을 넣으면 더 이해가 쉬울 것 같아서 문제에는 없지만 추가하였다.
```
